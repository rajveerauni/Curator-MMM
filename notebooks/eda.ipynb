# Curator MMM — Exploratory review

Quick sanity checks on the synthetic weekly panel before fitting.

```python
from pathlib import Path

import pandas as pd

ROOT = Path("..").resolve()
df = pd.read_csv(ROOT / "data" / "synthetic_mmm_data.csv", parse_dates=["week"])
df.head()
```

```python
summary = df[["tv_spend", "google_spend", "instagram_spend", "sales"]].describe()
summary
```

```python
ax = df.set_index("week")[["tv_spend", "google_spend", "instagram_spend"]].plot(
    figsize=(11, 4), title="Weekly spend by channel"
)
ax.set_ylabel("Spend ($)")
```
